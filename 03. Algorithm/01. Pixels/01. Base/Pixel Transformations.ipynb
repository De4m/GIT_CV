## Prepare

```python
im = cv2.imread("baboon.png")
img = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
#plt.imshow(im)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
plt.show()
```

## Histograms

```python
hist_b = cv2.calcHist([im],[0],None,[256],[0,256])
hist_g = cv2.calcHist([im],[1],None,[256],[0,256])
hist_r = cv2.calcHist([im],[2],None,[256],[0,256])
hist  = cv2.calcHist([img],[],None,[256],[0,256])

plt.figure(figsize=(10,10))
plt.title('Histograms')
plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
plt.subplot(2,3,2)
plt.imshow(img)
plt.subplot(2,3,3)
plt.plot(hist,color='black')
plt.subplot(2,3,4)
plt.plot(hist_b,color='blue')
plt.subplot(2,3,5)
plt.plot(hist_g,color='green')
plt.subplot(2,3,6)
plt.plot(hist_r,color='red')
plt.show()
```

## Brightness and Contrast Adjustments

$$ g(i,j) = \alpha f(i,j) - \beta $$
where:  
- $\alpha$ - simple contrast control
- $\beta$ - simple brightness control

```python
# prepare variable
a=1
b=0
pos=0

# Show modify contrast
plt.figure(figsize=(20,20))
for a in range(10,15,1):
    img2 = a/10*img + b      # modify contrast
    img2[img2>255] = 255     # clipping 
    img2 = np.uint8(img2)
    plt.subplot(2,5,1+pos)
    plt.imshow(img2)    

    hist  = cv2.calcHist([img2],[],None,[256],[0,256])
    plt.subplot(2,5,1+pos+5)
    plt.plot(hist,color='black')
    
    pos += 1
plt.show()
```

```python
# prepare variable
a=1
b=0
pos=0

# Show modify contrast
plt.figure(figsize=(20,20))
for b in range(0,150,10):
    img2 = a*img + b      # modify contrast
    img2[img2>255] = 255     # clipping 
    img2 = np.uint8(img2)
    plt.subplot(2,15,1+pos)
    plt.imshow(img2)    

    hist  = cv2.calcHist([img2],[],None,[256],[0,256])
    plt.subplot(2,15,1+pos+15)
    plt.plot(hist,color='black')
    
    pos += 1
plt.show()
```
