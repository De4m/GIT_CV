```python
import os
from PIL import Image
```

```python
# Base functions
def get_imlist(path):
    return [os.path.join(path,f) for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
pil_im = Image.open("./data/empire.jpg")
pil_im 
```

```python
pil_im.convert("L")
```

```python
get_imlist('./data')
```

```python
pil_im.thumbnail((128,128))
pil_im
```

```python
box = (100,100,400,400)
region = pil_im.crop(box)
region
```
