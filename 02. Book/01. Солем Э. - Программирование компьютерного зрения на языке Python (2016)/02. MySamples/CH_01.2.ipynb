```python
import os
from PIL import Image
from pylab import *
%matplotlib inline
```

```python
# Base functions
def get_imlist(path):
    return [os.path.join(path,f) for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
im = array (Image.open('./data/empire.jpg'))
imshow(im)
x = [100,100,400,400]
y = [200,500,200,500]
plot (x,y,'r*')
plot (x[:2],y[:2])
title('Title')
# axis('off')
show()
```

```python
im = array (Image.open('./data/empire.jpg').convert('L'))
figure()
gray()
contour (im,origin='image')
axis('equal')
axis('off')
```
