```python
im = array (Image.open('./data/empire.jpg').convert('L'))
imx = np.zeros(im.shape)
filters.sobel(im,1,imx)
imy = np.zeros(im.shape)
filters.sobel(im,0,imy)
magnitude = sqrt(imx**2+imy**2)

plt.figure()
plt.title('imx imy')
plt.gray()
plt.subplot(1,4,1)
plt.imshow(imx)
plt.subplot(1,4,2)
plt.imshow(imy)
plt.subplot(1,4,3)
plt.imshow(magnitude)

```
